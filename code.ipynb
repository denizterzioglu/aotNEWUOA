```python
import numpy as np
import pandas as pd

np.random.seed(1)
n = 70
lower = np.array([-2.5, -1.5])
upper = np.array([1.5, 2.5])
x = np.column_stack((np.random.uniform(lower[0], upper[0], size=n),
                     np.random.uniform(lower[1], upper[1], size=n)))
def f(x):
    return 20 + x[:,0]**2 + x[:,1]**2 - 10*(np.cos(2*np.pi*x[:,0]) + np.cos(2*np.pi*x[:,1]))
y = f(x)
```

```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, scale_color_gradient, guide_colorbar

df = pd.DataFrame({"x": x[:,0], "y": x[:,1]})

# Manuelle Erstellung einer Farbskala
my_colors = ["#440154", "#3E4A89", "#21908C", "#5DC863", "#FDE725"]

(ggplot(df, aes(x="x", y="y", color=y))
 + geom_point()
 + scale_color_gradient(low=my_colors[0], high=my_colors[-1], colors=my_colors[1:-1])
 , guide_colorbar(title="y"))
```

```python
from sklearn.svm import SVR
import numpy as np
import pandas as pd
from plotnine import ggplot, aes

model = SVR(kernel='rbf', gamma='scale')
model.fit(x, y)

nplot_dim = 100
xplot = pd.DataFrame(np.array(np.meshgrid(np.linspace(lower[0], upper[0], nplot_dim), 
                                           np.linspace(lower[1], upper[1], nplot_dim))).T.reshape(-1, 2), 
                     columns=['x.1', 'x.2'])
yplot = model.predict(xplot)
df = pd.concat([xplot, pd.DataFrame(yplot, columns=['y'])], axis=1)

(ggplot(data=df, mapping=aes(x='x.1', y='x.2', z='y')))
```
